```python
# Load shots
with open('./data/players_shots_clean.pickle', 'rb') as f:
    players_shots = pickle.load(f)
with open('./data/players_shots_density_restricted.pickle', 'rb') as f:
    players_shots_density = pickle.load(f)
with open('./data/players_shots_density_made_restricted.pickle', 'rb') as f:
    players_shots_density_made = pickle.load(f)
# with open('./data/players_shots_density.pickle', 'rb') as f:
#     players_shots_density = pickle.load(f)
# with open('./data/players_shots_density_made.pickle', 'rb') as f:
#     players_shots_density_made = pickle.load(f)
```

```python
# Plot the location of the shots
name = 'Stephen Curry'

fig, axs = plt.subplots(
    ncols=2, nrows=2,
    figsize=(8, 4),
    facecolor='white',
    edgecolor='black',
    dpi=300
)
gs = axs[0, 1].get_gridspec()
for ax in axs[:, 0]:
    ax.remove()
axbig = fig.add_subplot(gs[:, 0])
ShotCharts.shots_chart_points(axbig, players_shots, name, False)
axbig.set_aspect('equal')

ShotCharts.shots_chart(
    axs[0, 1], players_shots_density, name, 'Shots attempted density', False
)
axs[0, 1].set_aspect('equal')
ShotCharts.shots_chart(
    axs[1, 1], players_shots_density_made, name, 'Shots made density', False
)
axs[1, 1].set_aspect('equal')
fig.tight_layout()

plt.savefig('./figures/curry.eps', transparent=True, format='eps')
plt.show()
```
